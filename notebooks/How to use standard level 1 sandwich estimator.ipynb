Now we need to make a function that outputs, for a given subject, a list of the very-dictionary-like _Bunch_-objects with the following fields:

* conditions: list of condition for every trial
* onsets: list of onsets for every trial
* durations: list of durations for every condition/trial

```python
def get_subject_info(subject_id):
    import pandas
    from nipype.interfaces.base import Bunch

    events = pandas.read_csv('/home/gdholla1/data/birte/Designfiles/Duration4/%s.des' % subject_id, sep='\t')
    
    events['onset'] = events.onset.apply(lambda x: x.replace(',', '.') if type(x) == str else x)
    events['onset'] = events.onset.astype(float)
    
    events['duration'] = events.onset.apply(lambda x: x.replace(',', '.') if type(x) == str else x)
    events['duration'] = events.onset.astype(float)    
    
    info = [Bunch(conditions=events['%event'].unique().tolist(), 
              onsets=[e.onset.tolist() for g, e in events.groupby(['%event'])],
              durations=[e.duration.tolist() for g, e in events.groupby(['%event'])])]
    
    return info
```

Now we import some more nipype modules

```python
import nipype.interfaces.utility as util
import nipype.pipeline.engine as pe
import nipype.interfaces.io as nio
import nipype.algorithms.modelgen as model
```

```python
# Create identity node that iterates over different subjects
identity = pe.Node(util.IdentityInterface(fields=['subject_id']), name='identity')
identity.iterables = [('subject_id', ['as1t'])]


# Create a selector-node that gets corresonding functional runs
templates = {'func':'/home/gdholla1/data/in_limbo/birte/func/_subject_id_{subject_id}/_highpass0/*ii2_dtype_mcf_mask_smooth_mask_gms_tempfilt.nii.gz'}
selector = pe.Node(nio.SelectFiles(templates), name='selector')
selector.inputs.force_lists = True

# Put get_subject_info-function in a node using Function-interface
get_info_node = pe.Node(util.Function(function=get_subject_info,
                                      input_names=['subject_id'],
                                      output_names=['session_info']),
                        name='get_info_node')


# use SpecifyModel to create a session_info-object the
# Sandwich-interface (but also FSL-interface, btw) can deal
# with.
spec = pe.Node(model.SpecifyModel(), name='spec')
spec.inputs.high_pass_filter_cutoff = 128
spec.inputs.time_repetition = 2
spec.inputs.input_units = 'secs'

# Connect identity to file-Selector (functional data) 
# and subject_info_node (behavioral data)
sandwich_level1.connect(identity, 'subject_id', selector, 'subject_id')
sandwich_level1.connect(selector, 'func', spec, 'functional_runs')
sandwich_level1.connect(identity, 'subject_id', get_info_node, 'subject_id')
sandwich_level1.connect(get_info_node, 'session_info', spec, 'subject_info')

# Put functional data ...
sandwich_level1.connect(selector, 'func', sandwich_level1.get_node('inputspec'), 'functional_runs')

# ... and behavioral data into the sandwich-estimation workflow
sandwich_level1.connect(spec, 'session_info', sandwich_level1.get_node('inputspec'), 'session_info')

# Set up some contrasts
cont1 = ['Speed>acc','T', [1, 3],[1, -1]]
cont2 = ['Acc>speed','T', [1, 3],[-1, 1]]
cont3 = ['Acc','T', [1],[1]]
sandwich_level1.inputs.inputspec.contrasts = [cont1, cont2, cont3]

# Set up TR
sandwich_level1.inputs.inputspec.TR = 2.0
```
