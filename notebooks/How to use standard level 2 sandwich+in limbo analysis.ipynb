The field _z_threshold_ and _alpha_in_limbo_ refer to the z_threshold for the level 2 analysis and the alpha level of the in_limbo test

```python
level2.inputs.inputspec.z_threshold = 2.3
level2.inputs.inputspec.alpha_in_limbo = 0.05
```

Please note that *all the brain images should be in the same space* (probably MNI152).

This goes for the following inputs:
* copes
* residuals
* varcopes

I registered these files to MNI152 space using the [create_reg_workflow](http://www.mit.edu/~satra/nipype-nightly/interfaces/generated/nipype.workflows.fmri.fsl.preprocess.html#create-reg-workflow) from the standard Nipype library.

We now use a SelectFiles-interface to collect all the necessary files

```python
import nipype.pipeline.engine as pe
import nipype.interfaces.io as nio
import re, glob

reg = re.compile('.*/_subject_id_([0-9a-z]{4})/')
sids = [reg.match(fn).group(1) for fn in glob.glob('/home/gdholla1/data/in_limbo/birte/copes/_subject_id_*/')]

templates = {'cope':'copes_mni/{subject_id}/cope{contrast}.nii.gz',
             'varcope':'varcopes_mni/{subject_id}/varcope{contrast}.nii.gz',
            'residuals':'residuals_mni/{subject_id}/residuals_*.nii.gz',
            'design_matrices':'design_matrices/_subject_id_{subject_id}/design_matrix_cut*.npy',
            'dof':'dof/_subject_id_{subject_id}/dof',
            'contrast_vector':'contrast_vector/_subject_id_{subject_id}/contrast_vector{contrast}.nii.gz'}


selector = pe.MapNode(nio.SelectFiles(templates), 
                      iterfield=['subject_id'],
                      name='selector')
selector.inputs.subject_id = sids
selector.inputs.contrast = 1

selector.inputs.base_directory = '/home/gdholla1/data/in_limbo/birte/level1_sandwich/'
```

Connect all the  nodes and we're done

```python
level2.connect([(selector, 
                 level2.get_node('inputspec'),
                 [('cope', 'copes'),
                  ('varcope', 'varcopes'),
                  ('residuals', 'residuals'),
                  ('dof', 'dof_files'),
                  ('contrast_vector', 'contrast_vector'),
                  ('design_matrices', 'design_matrices')]
                 )])
```
